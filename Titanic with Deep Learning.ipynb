#### Importing Our  Packages

```python
import pandas as pd

import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

from sklearn.preprocessing import normalize, scale

import matplotlib.pyplot as plt
```

#### Reading Our File

```python
df=pd.read_csv('titanicnew.csv')
#I made EDA (Exploratory Data Analysis) in the csv file that I named titanic new.
```

```python
df.head()
```

```python
df.isnull().sum()
#We check if there are spaces in our columns.
#It is normal for the column named Survived to be empty. That column is the column we need to guess.
```

#### Preparing Data For The Model

```python
train=df[:891]
test=df[891:]
#We first combined our file as train and test. Now we separate it.
```

```python
x=train.drop('Survived',axis=1) 
y=train[['Survived']]
#We determine x and y.
```

```python
x.head()
#Let's check x
```

```python
y.head()
#And let's check y.
```

```python
df.shape
#We see how many columns and rows our data consists of.
```

#### Deep Learning

```python
model=Sequential()
model.add(Dense(150,activation='relu'))  
model.add(Dense(95,activation='relu'))   
model.add(Dense(65,activation='relu')) 
model.add(Dense(30,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1,activation='sigmoid'))  
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

#We write the layers, density and necessary parameters for deep learning.
```

```python
model.fit(x,y,epochs=250, validation_split=.20,batch_size=45,verbose=1)
#We train our model and see the success rate.
#Our success rate is accuracy for the classification model.
```

```python
scores=model.evaluate(x,y)
```

```python
model.summary()
#We see how many millions of parameters it works on.
```

#### Scaling

```python
x=scale(x)
#If we scale our data, our success rate will increase. We do scaling.
```

```python
history=model.fit(x,y,epochs=250,validation_split=.20,batch_size=45,verbose=1)
#After scaling, we train our model again.
#and we save the data we trained to the variable called history. We will use history to draw a chart.
```

```python
scores1=model.evaluate(x,y)
```

```python
plt.plot(history.history['accuracy'],label='Accuracy') 
plt.plot(history.history['val_accuracy'],label='Val Accuracy') 
plt.legend()
```
